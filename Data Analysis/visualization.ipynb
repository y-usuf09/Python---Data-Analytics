```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t,x) # for graph
```

```python
plt.plot(t,x, 'o-')
plt.plot(t,y, 'x-')
```

```python
plt.figure(figsize=(6,4))
plt.plot(t,x, 'ro--') # => r- red colour
plt.plot(t,y, 'h', markersize=10) # h=> hexagon
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x, marker='o', color='#00ff00',
         markersize=15, markeredgecolor='blue',
         markerfacecolor='yellow') 
plt.plot(t,y, 'h', markersize=25)
```

```python
years = [2010,2011,2012,2013,2014,2015]
prices = [120,90,140,150,180,130]
fig = plt.figure(figsize=(4,2))
ax1 = fig.add_axes([0,0,1,1])
ax1.pie(prices, labels=years,autopct='%.2f%%')
plt.show()
```

# Customization

```python
plt.figure(figsize=(4,2))
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
```

```python
plt.figure(figsize=(4,2), facecolor='orange')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
```

```python
plt.figure(figsize=(4,2), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines(140, 2010, 2015)
```

```python
plt.figure(figsize=(4,2), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines(140, 2010, 2015)
plt.vlines(2012,0,180)
```

```python
plt.figure(figsize=(4,2), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines([140,140], [2010, 2012], [2015,2015], linestyles=['dashed', 'solid'])
```

```python
plt.figure(figsize=(4,2), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines([140,140,150],
            [2010, 2012,2010], 
            [2015,2015,2013], 
            linestyles=['dashed', 'solid'])

```

```python
plt.figure(figsize=(4,2), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines([140,140,150],
            [2010, 2012,2010], 
            [2015,2015,2013], 
            linestyles=['dashed', 'solid'])
plt.vlines([2012,2013],
           [min(prices),min(prices)],
           [140,150],
           linestyles='dotted',
           colors=['red','blue'],
           label='Price')
```

```python
plt.figure(figsize=(6,4), facecolor='#ededed')
plt.plot(years,prices, 'ro--')
plt.title("Prices of stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines([140,140,150],
            [2010, 2012,2010], 
            [2015,2015,2013], 
            linestyles=['dashed', 'solid'])
plt.vlines([2012,2013],
           [min(prices),min(prices)],
           [140,150],
           linestyles='dotted',
           colors=['red','blue'],
           label='Price')
plt.text(2011, 142, 'Important Year', fontdict={'size':12})
plt.show()
```
