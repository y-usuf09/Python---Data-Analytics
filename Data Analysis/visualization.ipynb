```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t,x) # for graph
```

```python
plt.plot(t,x, 'o-')
plt.plot(t,y, 'x-')
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x, 'ro--') # r- red colour
plt.plot(t,y, 'h', markersize=10)
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x, marker='o', color='#00ff00',
         markersize=15, markeredgecolor='blue',
         markerfaceclor='yellow') 
plt.plot(t,y, 'h', markersize=25)
```
